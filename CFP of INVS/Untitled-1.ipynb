```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the extracted CSV file
df = pd.read_excel('university_carbon_footprint.xlsx')

# Add Addis Ababa Science and Technology University (AASTU) to the dataframe
aastu_data = {'University': 'Addis Ababa Science and Technology University', 'Country': 'ET', 'Annual CF (tCO2e)': 1135, 'CF per capita (tCO2e/pers)': 0.12}
df = pd.concat([df, pd.DataFrame([aastu_data])], ignore_index=True)

df["CF per capita (tCO2e/pers)"] = df["CF per capita (tCO2e/pers)"].apply(replace_str_with_zero)
df = df[df['CF per capita (tCO2e/pers)'] != 0]
df = df.sort_values('CF per capita (tCO2e/pers)')
# Plotting the data
plt.figure(figsize=(12, 8))

# Highlight AASTU in a different color
colors = ['blue' if uni != 'Addis Ababa Science and Technology University' else 'red' for uni in df['University']]

plt.barh(df['University'], df['CF per capita (tCO2e/pers)'], color=colors)
plt.xlabel('CF per capita (tCO2e/pers)')
plt.ylabel('University')
plt.title('Comparison of Annual Carbon Footprint (tCO2e) per capita Among Universities')
plt.tight_layout()
plt.show() 
```

```python
from logging import exception


def replace_str_with_zero(x):
    
    if isinstance(x, str):
        try:
            return float(x)
        except Exception as e:
            return 0
    return x
```
